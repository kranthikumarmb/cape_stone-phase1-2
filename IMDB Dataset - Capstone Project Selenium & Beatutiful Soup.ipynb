```python
from bs4 import BeautifulSoup

# Get the page source
page_source = driver.page_source

# Create a BeautifulSoup object
soup = BeautifulSoup(page_source, 'html.parser')

# Find all movie name elements
name_elements = soup.find_all('h3', class_='lister-item-header')

# Extract the movie names
movie_names = []
for element in name_elements:
    movie_name = element.a.text.replace('(', '').replace(')', '').replace('\n', '')
    movie_names.append(movie_name)

# Print the movie names
for name in movie_names:
    print(name)
```

```python
serial=[]
s_n1=soup.find_all('span',class_='lister-item-index unbold text-primary')
for span in s_n1:
    s_no=span.text.strip()
    s_no=s_no.strip('.')
    serial.append(s_no)
```

```python
duration=[]
du=soup.find_all('span',class_='runtime')
for k in du:
    dur=k.text.replace('min','')
    duration.append(dur)
a=print(','.join(duration))
```

```python
year=[]
yr=soup.find_all('span',class_='lister-item-year text-muted unbold')
for yrs in yr:
    yr=yrs.text.strip()
    yr=yr.strip('()').replace(',','')
    year.append(yr) 
```

```python
rating=[]
rat=soup.find_all('div',class_='inline-block ratings-imdb-rating')
for mn in rat:
    rt=mn.text.replace('\n','')
    rating.append(rt)
```

```python
metascore=[]
meta=soup.find_all('div',class_='lister-item-content')
for met in meta:
    movie_title= met.find('h3',class_='lister-item-header').find('a').text
    score_elem=met.find('div',class_='inline-block ratings-metascore')
    
    if score_elem is not None:
        mea=score_elem.text.replace('Metascore','').replace(',','').strip()
        if mea.isnumeric():
            metascore.append({'movie':movie_title,'metascore':mea})
        else:
            metascore.append({'movie':movie_title,'metascore':' '})
            
    else:
        metascore.append({'movie':movie_title,'metascore':' '})
metascore_values = [entry['metascore'] for entry in metascore]
for value in metascore_values:
    print(value)
```

```python
directors1 = []
directors2 = []
directors3 = []

direct_elems = soup.find_all('div', class_='lister-item-content')

for direct_elem in direct_elems:
    director_names = []
    director_elements = direct_elem.find_all('a')

    for element in director_elements:
        href = element.get('href')
        if href and 'adv_li_dr' in href:
            director_names.append(element.text.strip())

    # Extract the first director
    if director_names:
        director1 = director_names[0]
    else:
        director1 = ' '  # Fill with space if not available
    directors1.append(director1)

    # Extract the second director
    if len(director_names) > 1:
        director2 = director_names[1]
    else:
        director2 = ' '  # Fill with space if not available
    directors2.append(director2)

    # Extract the third director
    if len(director_names) > 2:
        director3 = director_names[2]
    else:
        director3 = ' '  # Fill with space if not available
    directors3.append(director3)

# Print the extracted directors' names
for i in range(len(directors1)):
    print(f"Director1: {directors1[i]}")
    print(f"Director2: {directors2[i]}")
    print(f"Director3: {directors3[i]}")
```

```python
directors_data = {
    's_no': serial,
    'movie':movie_names,
    'Director1': directors1,
    'Director2': directors2,
    'Director3': directors3,
    'Duration': duration,
    'year of release': year,
    'Rating': rating,
    'Meatscore': metascore_values
}

df = pd.DataFrame(directors_data)
df.insert(0, 's_no', df.pop('s_no'))  # Move 's_no' column to the first position
df.set_index('s_no', inplace=True)  # Set 's_no' column as the index
df
```

```python
stars1 = []
stars2 = []
stars3 = []
stars4 = []

for direct_elem in direct_elems:
    star_names = []
    star_elements = direct_elem.find_all('a')
    
    for element in star_elements:
        href = element.get('href')
        if href and 'adv_li_st' in href:
            star_names.append(element.text.strip())

    # Extract the first star
    if star_names:
        star1 = star_names[0]
    else:
        star1 = ' '  # Fill with space if not available
    stars1.append(star1)

    # Extract the second star
    if len(star_names) > 1:
        star2 = star_names[1]
    else:
        star2 = ' '  # Fill with space if not available
    stars2.append(star2)

    # Extract the third star
    if len(star_names) > 2:
        star3 = star_names[2]
    else:
        star3 = ' '  # Fill with space if not available
    stars3.append(star3)

    # Extract the fourth star
    if len(star_names) > 3:
        star4 = star_names[3]
    else:
        star4 = ' '  # Fill with space if not available
    stars4.append(star4)

# Create a DataFrame with the star names
stars_data = {
    'Star1': stars1,
    'Star2': stars2,
    'Star3': stars3,
    'Star4': stars4
}
```
